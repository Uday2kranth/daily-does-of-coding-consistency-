```python
class Node:
    def __init__(self, data):
        self.data = data
        self.link = None  # Instructor uses 'link', not 'next'
        return

class LinkedList:
    def __init__(self):
        self.head = None
        return

    def append(self, data):
        temp = Node(data)
        
        # Case 1: List is empty
        if self.head == None:
            self.head = temp
        
        # Case 2: List has elements
        else:
            last = self.head
            # Traverse to the last node
            while(last.link != None):
                last = last.link
            
            # Link the new node
            last.link = temp
            
        print("appended") #
        return

    def display(self):
        if self.head == None:
            print("List is empty")
        else:
            temp = self.head
            while(temp != None):
                print(temp.data)
                temp = temp.link
        return

    def delete_first(self):
        if self.head == None:
            print("List is empty")
        else:
            target = self.head # This is the node to be deleted
            
            # Move head to the next node
            self.head = self.head.link
            
            # Disconnect the target node
            target.link = None
            
            # Delete the object explicitly (Instructor style)
            del target
            print("First element deleted")
        return

# --- Driver Code (Based on Screenshot 5) ---
obj = LinkedList()

print("--- Appending Elements ---")
obj.append(10)
obj.append(20)
obj.append(30)

print("\n--- Displaying List ---")
obj.display()

print("\n--- Appending More ---")
obj.append(40)
obj.append(50)

print("\n--- Displaying List ---")
obj.display()

print("\n--- Deleting First Element ---")
obj.delete_first()
obj.display()
```
