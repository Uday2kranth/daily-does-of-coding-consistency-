- file handling is used to store data permanently in a file and retrive it when required

### 1.Opening a file 
- python provides the built-in function open()


```python
file_object=open('test.txt','a')
```

- **filename** ------> name of the file (e.g 'data.txt')
- **Mode** ---------> defnes what we want to do with the file

### common modes of operations we can perform in the file handlings are 
**Mode:** **Descriptions**
- 'r'  read(default). Error if file does not exits
- 'w'  write,create new file or overwrites  existing one 
- 'a' append.add the data at the end of the file 
- 'x' create. Create a new file. error if file already exists

### writing files 
- if file does not exist ,create one 
- if file exist -it **over writes** (in 'w' mode)

```python
f=open('name.txt','w') # 'w' mode opens a file for writing. If the file already exists, it truncates the file to zero length. If the file does not exist, it creates a new file.
f.write('gagan\n')
f.write('saikiran\n')
f. write('sai ram\n')
f.write('Tharun\n')
f.write('akash\n')
f.close()
```
